# Voronoi Tessellations

```python
#Importing libraries
import numpy as np
%pylab inline
import matplotlib.pyplot as plt
import voronoi as vorn
```

## Plotting a function over a Voronoi Meshes using class voronoifield

```python
P = np.random.random((3000,2))
X,Y = P[:,0],P[:,1]
F = abs(np.sin(10*(X+Y**3)))

#Defining voronoifield object
field = vorn.voronoifield(X,Y,F)
#Constructing mesh 2d
field.voronoimesh2d(condition="finite", Lx=[0,1], Ly=[0,1])
```

```python
fig = plt.figure( figsize = (10,10) )
axe = fig.add_subplot(1,1,1)
field.voronoishow( cmap="jet", axes = axe )
```

## Plotting the area function of each Voronoi cell

```python
P = np.random.random((200,2))
X,Y = P[:,0],P[:,1]
#Defining voronoifield object
field = vorn.voronoifield(X,Y)
#Constructing mesh 2d
field.voronoimesh2d(condition="finite", Lx=[0,1], Ly=[0,1])
#Calculating area of each cell
field.voronoisurface()
field.f = field.surface
```

```python
fig = plt.figure( figsize = (10,10) )
axe = fig.add_subplot(1,1,1)
field.voronoishow( cmap="jet", axes = axe )
```

## Testing in Bolshoi

```python
#Loading data of Bolshoi
datos = np.loadtxt( "fof_bolshoi.dat", usecols=(1,2,3,8) )
```

```python
#Filtering data
zmin = 30.0
zmax = 40.0
X, Y, Z, F = datos[ (zmin<=datos[:,2])&(datos[:,2]<=zmax) ].T
```

```python
#Defining voronoifield object
field = vorn.voronoifield(X,Y)
#Constructing mesh 2d
field.voronoimesh2d(condition="finite", Lx=[0,250], Ly=[0,250])
#Calculating area of each cell
field.voronoisurface()
#Density
field.f = np.log10(F/field.surface)
```

```python
#Plotting
fig = plt.figure( figsize = (18,8) )
axe1 = fig.add_subplot(1,2,1)
field.voronoishow( cmap="jet", axes = axe1, lw = 0.1 )
axe2 = fig.add_subplot(1,2,2)
axe2.plot( X, Y, ".", color="black", ms=1.0 )
```
