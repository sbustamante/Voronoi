# Differentiation with Voronoi Tessellations

```python
#Importing libraries
import numpy as np
%pylab inline
import matplotlib.pyplot as plt
import voronoi as vorn
```

## Functions for Voronoi Meshes using class voronoifield

```python
P = np.random.random((3000,2))
X,Y = P[:,0],P[:,1]
F = abs(np.sin(10*(X+Y**3)))

field = vorn.voronoifield(X,Y,F)
field.voronoimesh2d(condition="finite", Lx=[0,1], Ly=[0,1])
```

```python
fig = plt.figure( figsize = (10,10) )
axe = fig.add_subplot(1,1,1)
field.voronoishow( cmap="jet", axes = axe )
```
