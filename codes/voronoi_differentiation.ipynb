# Differentiation with Voronoi Tessellations

```python
#Importing libraries
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import collections
from scipy.spatial import Voronoi, voronoi_plot_2d, Delaunay, KDTree
%pylab inline
```

## Second way: using directly Voronoi tessellations

```python
P = np.random.random((1000,2))
X,Y = P[:,0],P[:,1]
Pf = abs(np.sin(10*X*Y))
vor = Voronoi(P)
#voronoi_plot_2d(vor)
```

```python
fig = plt.figure(figsize=(10,10))
axes1 = fig.add_subplot(1,1,1)
#plt.scatter(X, Y, marker='o', s = size, linewidths = 1.5, color = "red", edgecolor='k', zorder = 4)
axes1.set_xlim( (0,1) )
axes1.set_ylim( (0,1) )

i=0
for reg in np.array(vor.regions)[vor.point_region]:
    try:
        if sum(np.array(reg)<0)==0:
            poly = vor.vertices[reg]
            p = matplotlib.patches.Polygon(poly, facecolor=(0,0,0.2+0.8*Pf[i]/np.max(Pf)),alpha=1)
            axes1.add_patch(p)
    except:
        None
    i+=1
```
